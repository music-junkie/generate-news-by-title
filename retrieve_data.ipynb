```python
import json
import requests

import time

import pandas as pd
import numpy as np
```

```python
from IPython.display import display, clear_output

result_dict = []
for page in range(1, 10000):
        
    payload = {'chrono': 'news', 'locale': 'ru', 'page': page, 'per_page':'24'}    
    url = 'https://meduza.io/api/v3/search'

    response = requests.get(url, params=payload)    

    for i in response.json()['documents']:
        result_dict.append(response.json()['documents'][i])

    time.sleep(2)

    clear_output(wait=True)
    display(f'Page: {page}, count: {len(result_dict)}')

    if response.json()['has_next'] != True:
        break
    
```

```python
with open('data/data.json', 'w') as outfile:
    json.dump(result_dict, outfile)
```

```python
df = pd.read_json('data/data.json')
```
